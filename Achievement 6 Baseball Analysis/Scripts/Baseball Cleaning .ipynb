# 11 Create a Region Column

Now that I have a location tag, I think that it would be beneficial to my analysis to have a region tag based on the team abbreviation.  

```python
# find out the differnt locations 
CleanBBData['Location'].value_counts()
```

```python
# create a list of regions 
Southeast = ['TBR', 'ATL', 'MIA', 'WSN']
```

```python
Southwest = ['ARI', 'TEX', 'HOU']
```

```python
West = ['SFG', 'COL', 'SDP', 'LAD', 'LAA', 'SEA', 'OAK']
```

```python
Midwest = ['MIL', 'CHC', 'CIN', 'STL', 'DET', 'MIN', 'CLE', 'CHW', 'KCR']
```

```python
Northeast = ['PIT', 'NYM', 'TOR', 'BOS', 'PHI', 'BAL', 'NYY']
```

```python
# loop for region columns 
result=[]
for Location in CleanBBData['Location']:
    if Location in Northeast:
        result.append('Northeast')
    elif Location in Midwest:
        result.append('Midwest')
    elif Location in Southeast:
        result.append('Southeast')
    elif Location in Southwest:
        result.append('Southwest')
    else:
        result.append('West')
```

```python
# create column
CleanBBData['Game Region']=result
```

```python
# check
CleanBBData['Game Region'].value_counts()
```
